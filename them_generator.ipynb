```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
image = matplotlib.image.imread('/content/pic.jpg')
```

```python
X = image.reshape(-1,3)
X.shape
```

```python
X = X[:,:3]
```

```python
X
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
from PIL import Image, ImageDraw

def create_color_palatte(dominant_colors, palatte_size = (300,50)):
  palatte = Image.new("RGB", palatte_size)
  draw = ImageDraw.Draw(palatte)

  swatch_width = palatte_size[0] //  len(dominant_colors)

  for i, color in enumerate(dominant_colors):
    draw.rectangle([i * swatch_width, 0, (i + 1) * swatch_width, palatte_size[1]], fill= tuple(color))

  return palatte
```

```python
create_color_palatte(kmeans.cluster_centers_.astype(int))
```
